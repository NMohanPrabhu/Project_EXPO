```python
# Step 2: Import libraries
import pandas as pd
from pycaret.classification import *
```

```python
df=pd.read_csv("heart.csv")
```

```python
df
```

```python
# Step 4: Initialize PyCaret setup
clf = setup(
    data=df, 
    target=df['target'], 
    session_id=42, 
    normalize=True, 
    )
```

```python
# Step 5: Compare models (AutoML)
best_model = compare_models()
```

```python
#Analyze Model
evaluate_model(best_model)
```

```python
#Save Model
save_model(best_model, 'heart_pipeline')
```

```python
# load pipeline
loaded_best_pipeline = load_model('heart_pipeline')
loaded_best_pipeline
```
